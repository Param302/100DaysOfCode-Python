### Calculate `median` over `mean` if there are outliers in the data
- mean is the average of numbers in the list.
- median is the middle number in the sorted list.
- **Outlier** is some data point that differs from other observations.
- Outlier is the specific data (can be high or low data) relative to the rest of the data.
Like some persons have entered fake (high) salary, which can effect our rest of the data.
So mean will include that salary while calculated.
But, median doesn't include it.

### Use `agg()` method to do more than one operations

```python
country_grp["Salary"].agg(["mean", "median", "count", "min", "max"])
```

```python
# Getting Aggregate values of a specific country
country_grp["Salary"].agg(["mean", "median", "count", "min", "max"]).loc["India"]
```

```python
# Use `str.contains` to find a specific value in a series
df[df["Country"]=="India"]["LanguageHaveWorkedWith"].str.contains("Python")
```

```python
# Counting total number of Indian Developers who know Python
df[df["Country"]=="India"]["LanguageHaveWorkedWith"].str.contains("Python").sum()
```

### Using `apply()` on SeriesGroupby object, to apply a specific function

```python
country_grp["LanguageHaveWorkedWith"].apply(lambda series: series.str.contains("Python").sum()).loc["India"]
```

#### What percentage of people in each country knows Python

```python
no_of_response = df["Country"].value_counts()
no_of_response
```

```python
country_use_python = country_grp["LanguageHaveWorkedWith"].apply(lambda series: series.str.contains("Python").sum())
country_use_python
```

```python
python_df = pd.concat([no_of_response, country_use_python], axis="columns")
python_df
```

```python
python_df.rename(columns={"Country": "TotalRespondents", "LanguageHaveWorkedWith": "KnowsPython"}, inplace=True)
python_df
```

```python
# Calculating Percentage of dev knows Python
python_df["Percentage"] = round(python_df.KnowsPython / python_df.TotalRespondents * 100, 2)
```

```python
python_df.sort_values(by=["Percentage", "TotalRespondents"], ascending=[False, False], inplace=True)
python_df
```
