## JSON
### Use `to_json()` to export data in a JSON format
- pass `"records"` in `orient` parameter to make reocrds in list-like
- set `lines=True` parameter to set each record in a line.
- pass a number of spaces in `indent` parameter to indent each record

```python
india_df.to_json("./data/india_df_modified.json", indent=4)
```

## MySQL
- Installed `SQLAlchemy` package to read/write mysql data for pandas
- Installed `mysql-connector-python` to make connection with mysql.
- - mysql-connector-python is giving MySQLInterfaceError, it is not fully compatible with sqlalchemy so using `pymysql` instead.
- Install `pymysql` to make connection with mysql.

```python
from sqlalchemy import create_engine
import pymysql
```

```python
# mysql+pymysql://<username>:<password>@<host>:[port]/<dbname>[?<options>]
engine = create_engine("mysql+pymysql://root:password@localhost:3306/pandas_tut")
```

## Use `to_sql()` to add data to an sql table (also create engine with sqlalchemy first)
- `name` parameter takes the table name
- `con` parameter takes the connection/engine
- `if_exists` parameter gives error when it is "fail"(default) means table is already existed.
- - Set `if_exists="replace"` to drop the table and create a new table and insert values.
- - Set `if_exists="append"` to add new values in the existed table.

```python
# Writing data to mysql table
# return number of numbers are added
india_df.to_sql("india_stackoverflow_21", engine, if_exists="replace")
```

### Use `read_sql()` to read data from an sql table

```python
india_sql_df = pd.read_sql("india_stackoverflow_21", engine, index_col="ResponseId")
india_sql_df.head()
```

### Use `read_sql_query()` to read data from sql based on QUERY

```python
india_sql_df2 = pd.read_sql_query("SELECT * FROM india_stackoverflow_21", engine, index_col="ResponseId")
india_sql_df2.head()
```
