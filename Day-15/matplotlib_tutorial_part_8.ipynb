- Part 8: Plotting Time Series Data

```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt, dates as mpl_dates
```

```python
plt.style.use("seaborn")
```

```python
dates = [
    datetime(2022, 6, 11),
    datetime(2022, 6, 12),
    datetime(2022, 6, 13),
    datetime(2022, 6, 14),
    datetime(2022, 6, 15),
    datetime(2022, 6, 16),
    datetime(2022, 6, 17)
]

y = [0, 1, 3, 4, 6, 5, 7]
```

### Use `plt.plot_date()` to plot time series line plot
- set `"solid"` to `linestyle` parameter to make line visible

```python
plt.plot_date(dates, y, linestyle="solid")
plt.tight_layout()
```

#### Use `plt.gcf().autofmt_xdate()` to format dates in x-axis of current figure.
- gcf = get current figure

```python
plt.plot_date(dates, y, linestyle="solid")
plt.tight_layout()
plt.gcf().autofmt_xdate()
```

### Use `dates.DateFormatter()` to make format for dates
- Pass date format in it, same as [datetime](https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior) format

#### Use `plt.gca()` to get x-axis & y-axis
- For respective axis, use `set_major_formatter()` method and pass it in the date format.

```python
plt.plot_date(dates, y, linestyle="solid")
plt.tight_layout()
plt.gcf().autofmt_xdate()
date_format = mpl_dates.DateFormatter("%b, %d %Y")
plt.gca().xaxis.set_major_formatter(date_format)
```

### Loading `Bitcoin price` csv file of 2 weeks for plotting it

```python
data = pd.read_csv("./data/bitcoin_price.csv", parse_dates=["Date"])
price_date = data["Date"]
price_close = data["Close"]
```

```python
data.info()
```

```python
plt.title("Bitcoin Prices (USD)")
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.plot_date(price_date, price_close, fmt="o--", c="#00CC00", linewidth=3, markersize=10, label="Bitcoin Price")
plt.legend(loc="upper left")
plt.tight_layout()
plt.gcf().autofmt_xdate()
```
