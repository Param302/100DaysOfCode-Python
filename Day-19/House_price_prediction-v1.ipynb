```python
# Importing required libraries
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

## Importing data

```python
house_data = pd.read_csv("../input/home-data-for-ml-course/train.csv", index_col="Id")
test_data = pd.read_csv("../input/home-data-for-ml-course/test.csv", index_col="Id")

house_data.columns
```

### Selecting useful features

```python
useful_features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd', 'KitchenAbvGr', 'Fireplaces', 'PoolArea']
X = house_data[useful_features]
test_X = test_data[useful_features]
y = house_data["SalePrice"]
X.head()
```

```python
# All selected features are good to go
X.isnull().sum()
```

### Splitting `training` & `validating` data

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=30)
```

### Training the model

```python
my_model = RandomForestRegressor(random_state=30)
my_model.fit(train_X, train_y)
```

```python
val_preds = my_model.predict(val_X)
mae = mean_absolute_error(val_y, val_preds)
print(f"MAE: {mae}")
```
